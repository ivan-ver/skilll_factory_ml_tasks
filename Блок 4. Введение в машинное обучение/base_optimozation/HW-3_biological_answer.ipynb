```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import linear_model
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn import tree
from sklearn import metrics
from hyperopt import hp, fmin, tpe, Trials

from sklearn.model_selection import train_test_split 
import optuna
```

```python
data_set = pd.read_csv('./data/_train_sem09 (1).csv')
```

```python
data_set.head()
```

```python
X = data_set.drop('Activity', axis=1)
y = data_set['Activity']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.2)
```

```python
LR_param_grid = {'penalty': ['l2', 'none'] ,
              'solver': ['lbfgs', 'sag', 'liblinear', 'saga'],
              'C': [0.01, 0.1, 0.3, 0.5, 0.7, 0.9, 1]
              }
```

```python
DTC_param_grid = {'min_samples_leaf': list(np.linspace(5, 100, 50, dtype=int)),
              'max_depth': list(np.linspace(1, 30, 50, dtype=int)),
              'criterion':['entropy','gini']
              }
```

GridSearchCV

```python

grid_search_LR = GridSearchCV(
    estimator=linear_model.LogisticRegression(
        random_state=1, 
        max_iter=1000
    ), 
    param_grid=LR_param_grid, 
    cv=5, 
    n_jobs = -1
)  
```

```python
grid_search_LR.fit(X_train, y_train) 
y_test_pred_LR_grid_search = grid_search_LR.predict(X_test)
```

```python
            
grid_search_DTC = GridSearchCV(
    estimator=tree.DecisionTreeClassifier(
        random_state=42
    ), 
    param_grid=DTC_param_grid, 
    cv=5, 
    n_jobs = -1
)  
```

```python
grid_search_DTC.fit(X_train, y_train) 
y_test_pred_DTC_grid_search = grid_search_DTC.predict(X_test)
```

RandomizedSearchCV

```python
random_search_LR = RandomizedSearchCV(
    estimator=linear_model.LogisticRegression(
        random_state=1, 
        max_iter=1000
    ), 
    param_distributions=LR_param_grid, 
    cv=5,
    n_iter = 50, 
    n_jobs = -1
)  
```

```python
random_search_LR.fit(X_train, y_train) 
y_test_pred_LR_random_search = random_search_LR.predict(X_test)
```

```python
random_search_DTC = RandomizedSearchCV(
    estimator=tree.DecisionTreeClassifier(
        random_state=42
    ), 
    param_distributions=DTC_param_grid, 
    cv=5, 
    n_jobs = -1
)  
```

```python
random_search_DTC.fit(X_train, y_train) 
y_test_pred_DTC_random_search = random_search_DTC.predict(X_test)
```

Hyperopt

```python
space={'max_depth' : hp.quniform('max_depth', 15, 26, 1),
       'min_samples_leaf': hp.quniform('min_samples_leaf', 2, 10, 1)
      }
```

```python
def hyperopt_rf_DTC(params, cv=5, X=X_train, y=y_train):
    params = {
      'max_depth': int(params['max_depth']),
      'min_samples_leaf': int(params['max_depth'])
      }
    model = tree.DecisionTreeClassifier(**params, random_state=42)
    model.fit(X, y)
    score = metrics.f1_score(y, model.predict(X))
    return -score
```

```python
trials = Trials() 
best=fmin(hyperopt_rf_DTC,
          space=space,
          algo=tpe.suggest,
          max_evals=20,
          trials=trials
        )
print("Наилучшие значения гиперпараметров {}".format(best))
```

```python
model_DTC = tree.DecisionTreeClassifier(
    random_state=42, 
    max_depth=int(best['max_depth']),
    min_samples_leaf=int(best['min_samples_leaf'])
)
model_DTC.fit(X_train, y_train)
y_test_pred_DTC_hyperopt = model_DTC.predict(X_test)
```
